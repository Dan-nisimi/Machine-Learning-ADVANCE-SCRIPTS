# ADVANCE NUMPY OPERATIONS

```python
#Broadcasting in Numpy
# Rules of broadcasting are as follows:
# dimensions are aligned from the right
#dimensions is compatible if :
# 1. are equal (matches the other arrays dimensions)
# 2. one of the dimensions is 1

import numpy as np

#Array and scalar broadcasting

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr + 2)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
vector = np.array([1, 2, 3])
print(matrix + vector)

#Aggregation functions : Its commonly used to calculate the sum, mean, min, max, standard deviation, etc for arrays
print("sum: ", np.sum(arr))
print("mean: ", np.mean(arr))
print("min: ", np.min(arr))
print("max: ", np.max(arr))
print("std: ", np.std(arr))
print("sum along columns:", np.sum(arr, axis=0)) # axis=0 means sum along columns
print("sum along rows:", np.sum(arr, axis=1)) # axis=1 means sum along rows


#Boolean Indexing : Used to select elements from an array based on a condition
arr = np.array([1, 2, 3, 4, 5])
evens = arr[arr % 2 == 0]
print("Evens: ", evens)

arr[arr > 3] = 0 # replace values greater than 3 with 0
print("Modified Array: ", arr)

#Random Number Generation: Used to generate random numbers

#Setting random seeds
np.random.seed(42) # 42 is the seed value which means the random numbers will be the same every time

random_array = np.random.rand(3, 3)
print("Random Array: \n", random_array)

random_integers = np.random.randint(0, 10, size=(2,3)) # size=(2,3) means 2 rows and 3 columns
print("Random Integers: \n", random_integers)

```

```python
#Task 1 : Perform data cleaning, aggregation and filtering

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Inspect Data
print(df.info())
print(df.describe())

# Handle missing values
df["Age"] = df["Age"].fillna(df["Age"].median()) #fillna is used to fill missing values
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0]) #By using [0],
#we ensure that we are using only the first mode to fill in missing values.

# Remove duplicates
df = df.drop_duplicates()

# Filter data: Passengers in first class
first_class = df[df["Pclass"] == 1]
print("First Class Passengers: \n", first_class.head())

#Task 2: Generate visualizations to illustrate key insights

#Bar Chart: Survival rate by class
survival_by_class = df.groupby("Pclass")["Survived"].mean()
survival_by_class.plot(kind="bar", color="skyblue")
plt.title("Survival Rate by class")
plt.ylabel("Survival Rate")
plt.show()

# Histogram: Age distribution
sns.histplot(df["Age"], kde=True, bins=20, color="purple") #kde is used to add a kernel density
#estimate to the histogram. KDE is a smooth curve that represents the density of the data.
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot: Age vs Fare
plt.scatter(df["Age"], df["Fare"], alpha=0.5, color="green") #alpha is used to adjust the transparency
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

#Identify and interpret patterns or anomalies

#Summarize findings in a report
```
